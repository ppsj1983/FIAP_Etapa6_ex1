```python
import os
import subprocess

def get_latest_train_run_folder():
    train_path = 'yolov5/runs/train'
    if not os.path.exists(train_path):
        return None

    # Lista apenas pastas, ignorando ficheiros
    subfolders = [f.path for f in os.scandir(train_path) if f.is_dir()]

    if not subfolders:
        return None

    # Retorna a pasta com a data de criação mais recente
    return max(subfolders, key=os.path.getctime)

latest_run = get_latest_train_run_folder()

if latest_run:
    weights_path = os.path.join(latest_run, 'weights/best.pt')

    # Verifica se o ficheiro de pesos existe antes de correr
    if os.path.exists(weights_path):
        command = (
            f"python yolov5/detect.py "
            f"--weights {weights_path} "
            f"--img 640 "
            f"--source drive/MyDrive/FIAP/FIAP_NEXT/2026_b/test/ "
            f"--data yolov5/data/bolsa.yaml"
        )

        result = subprocess.run(command, shell=True, capture_output=True, text=True)

        print("Saída:\n", result.stdout)
        if result.stderr:
            print("Erros:\n", result.stderr)
    else:
        print(f"Pesos não encontrados em: {weights_path}")
else:
    print("Não foi possível encontrar a pasta de treinamento mais recente.")

```
